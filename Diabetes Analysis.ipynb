```python
#Exploratory Data Analysis
#Importing visualization libraries
import matplotlib.pyplot as plt
```

```python
#Univariate Visualization
```

```python
x = df['Glucose']
y = df['BMI']
z = df['Outcome']
plt.title('Histogram For Glucose')
plt.xlabel('Glucose Level')
plt.ylabel('Values')
plt.hist(x)
plt.show()
```

```python
#plotting hostogram for y
plt.hist(y)
plt.title('Histogram For BMI')
plt.xlabel('BMI')
plt.ylabel('Values')
plt.show()
```

```python
#Bivariate Visualization
```

```python
plt.scatter(x, y)
plt.title('Scatter Plot For Glucose & BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
```

```python
colors = {1 : 'red', 0:'green'}
plt.scatter(x, y,  c= z.apply(lambda x: colors[x]))
plt.title('Scatter Plot For Glucose & BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
```

The Above scatter plot shows if Glucose Value is greater than 135 and BMI is greater than 25, a woman might be Diabtic

```python
#Analyzing considering Gucose Pregnency and Age
x = df['Pregnancies']
y = df['Age']
z = df['Outcome']
plt.scatter(x, y,  c= z.apply(lambda x: colors[x]))
plt.title('Scatter Plot For Pregnancies & Age')
plt.xlabel('Pregnancies')
plt.ylabel('Age')
plt.show()
```

There doesn't seem to be any co-relation between Age & Number of pregnancies with Diabetes

```python
df.corr()
```

Above Correlation Matrix shows moderate correlation between Glucose values with the outcome, which is Diabetes. Outcome has lower correlation with number of Pregnancies, Age and BMI

```python
#Analyzing considering Glucose and Age
x = df['Glucose']
y = df['Age']
z = df['Outcome']
plt.scatter(x, y,  c= z.apply(lambda x: colors[x]))
plt.title('Scatter Plot For Glucose & Age')
plt.xlabel('Glucose')
plt.ylabel('Age')
plt.show()
```

The above scatterplot shows women with Glucose levels above 135 may have diabetes even though their age is as low as 22 years age. It also shows women above the age of 40 may have higher posibilities of diabetes even if their glucose level is 150 or above. Women with glucose level 175 and above might be definitely diabetic irrespective of their age.
