```python
#Bin variables
import numpy as np
df['GlucBin'] = np.where(df['Glucose'] <=135,'<=135', (np.where(df['Glucose'] >135 ,'>135', '0')))
```

```python
df.head()
```

Rule - Glucose > 135 is Diabetic and Glucose <= 135 is Non-Diabetic

```python
pd.pivot_table(df, values="Glucose", index=["GlucBin"], columns="Outcome", aggfunc = "count",fill_value=0)
```

Glucose <= 135 are nondiabetic. Accuracy level 77%. 
Glucose > 135 are Diabetic. Accuracy level 65%

```python
#Bin variables with Rule Glucose > 140 is diabetic
import numpy as np
df['GlucBin'] = np.where(df['Glucose'] <=140,'<=140', (np.where(df['Glucose'] >140 ,'>140', '0')))
df.head()
```

```python
#Pivot table with binned data
pd.pivot_table(df, values="Glucose", index=["GlucBin"], columns="Outcome", aggfunc = "count",fill_value=0)
```

Glucose <= 140 are nondiabetic. Accuracy level 76%. Glucose > 140 are Diabetic. Accuracy level 69%
