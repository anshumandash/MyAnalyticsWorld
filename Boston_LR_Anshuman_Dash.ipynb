```python
import numpy as np
import pandas as pd
```

```python
#creating a data frame using dictionary
mydict = {"Income": (50,25,15,10,np.NAN,np.NAN,20,10,np.NAN,999), "Age":(19,20,30,np.NAN, np.NAN, 25,20,52,46,9999)}
```

```python
mydict
```

```python
#Convert the dictionary into Pandas Dataframe
mydata = pd.DataFrame(mydict)
```

```python
mydata
```

```python
#Replace 9999,999 with NAN
mydata = mydata.replace((9999,999), np.NAN)
mydata
```

```python
#Delete the whole row which contains NAN, this works for any cell that has NAN
new = mydata.dropna(how = "any")
new
```

```python
# Drop the rows where all the values in a row are NAN, else don't delete the rows
new = mydata.dropna(how="all")
new
```

```python
# Replace the missing values or NANs with mean of the column
new = np.round(mydata.fillna(mydata.mean()),1)
#Inthis case, we are rounding off the mean value and the 1 at the end of the command means rounding off to 1 decimal
new
```

```python
# Replace the missing values or NANs with median of the column
new = np.round(mydata.fillna(mydata.median()),1)
#Inthis case, we are rounding off the mean value and the 1 at the end of the command means rounding off to 1 decimal
new
```

```python
# replace missing values with fixed values. This fixed value can be identified based on experience
new = mydata.fillna({"Age":30, "Income":15})
new
```

```python
#Replacing missing values with mean or median
#If the data has outliers, replace the missing values with median
#If the data is normal with no outliers, use mean values to replace 
```

```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Print many statements at the same time using this command
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
#import data sets and create a data frame
from sklearn import datasets
boston = datasets.load_boston()
```

```python
print(boston.DESCR)
```

```python
# Boston dataset features display
boston.keys()
boston.data.shape
boston.feature_names
boston.target
```

```python
# Create a data frame
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df.head()
```

```python
# Add dependent variable
boston_df['House_price'] = boston.target
boston_df.head()
boston_df.describe()
```

```python
# correlation matrix
x = boston_df.corr()
x
```

Creating features and labels and running correlation

```python
plt.subplots(figsize=(20,20))
sns.heatmap(x, cmap = 'RdYlGn', annot = True)
plt.show()
```

```python
# create features and lanbels for the normalized data
x = boston_df.drop('House_price', axis = 1)
y = boston_df['House_price']
x.head()
y.head()
```

```python
#Create train, test data with 75% and 25% split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.25, random_state = 1)
train_x.shape
test_x.shape
train_y.shape
test_y.shape
```

```python
# Let's import the regression object and define model
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm
```

```python
# Fit a model on the train data
lm.fit(train_x, train_y)
```

```python
# Evaluate the model
predict_test = lm.predict(test_x)
```

```python
#R2 Value
print("RSquare value for TEST data is - ")
np.round(lm.score(test_x, test_y)*100, 0)
print("RSquare value for TRAIN data is - ")
np.round(lm.score(train_x, train_y)*100, 0)
```

```python
# Predict on test and training data
predict_test = lm.predict(test_x)
```

```python
# Print the loss function  - MSE
import numpy as np
from sklearn import metrics
print("Mean Squaere Error(MSE) for Test Data is ")
np.round(metrics.mean_squared_error(test_y, predict_test), 0)
```

```python
from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error  (MAE) for Test Data is ")
np.round(mean_absolute_error(test_y, predict_test), 0)
```

```python
#Append Data
fdf = pd.concat([test_x, test_y], 1)
fdf['Predicted'] = np.round(predict_test, 1)
fdf['Predicted_Error'] = fdf['House_price'] - fdf['Predicted']
fdf
```

```python
plt.subplots(figsize = (20,20))
plt.scatter(fdf.House_price, fdf.Predicted_Error, color = 'red')
plt.xlabel ('House Price')
plt.ylabel ('Error')
plt.show()
```

```python
# FEATURE SELECTION - IMPORTANCE
ft_importances = pd.Series(lm.coef_,index=x.columns)
ft_importances.plot(kind='barh')
plt.show
```
